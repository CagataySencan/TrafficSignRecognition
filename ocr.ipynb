```python
import numpy as np
import cv2
```

```python
img = cv2.imread('starter.jpg')
#Alternatively: can be skipped if you have a Blackwhite image
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gray, img_bin = cv2.threshold(gray,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
gray = cv2.bitwise_not(img_bin)
```

```python
kernel = np.ones((2, 1), np.uint8)
img = cv2.erode(gray, kernel, iterations=1)
img = cv2.dilate(img, kernel, iterations=1)
out_below = pytesseract.image_to_string(img)
print("OUTPUT:", out_below)
```

```python
import requests
from PIL import ImageFilter
 
def process_image(url):
    image = cv2.imread(url)
    filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    # Applying cv2.filter2D function on the image
    sharpen_img=cv2.filter2D(image,-1,filter)
    gray = cv2.cvtColor(sharpen_img, cv2.COLOR_BGR2GRAY)
    return pytesseract.image_to_string(gray)

str = process_image("starter.jpg")
print(str)
```
